```python
df = pd.read_csv("/content/drive/MyDrive/Data mining/project 2/heart_disease (1).csv")
df
```

```python
import pandas as pd

df_encoded = pd.get_dummies(data=df, columns=df.columns.to_list())

df_encoded
```

```python
from mlxtend.frequent_patterns import apriori

# Generate frequent itemsets with minimum support of 0.3
frequent_itemsets = apriori(df_encoded, min_support=0.3, use_colnames=True)

# Display the result
print(frequent_itemsets)
```

```python
for item in frequent_itemsets.iloc[90]:
  print(item)
```

```python
from mlxtend.frequent_patterns import association_rules

total_itemsets = len(frequent_itemsets)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

```python
rules = rules.sort_values(by=['confidence', 'lift'], ascending=[False, False])

# Showing the first 10 rules
top_rules = rules[:10]
top_rules
```

```python
sorted_rules = rules.sort_values(['confidence', 'lift'], ascending=False)
top_10_rules = sorted_rules.iloc[:10]


print(top_10_rules.loc[:, ['confidence', 'lift']])
```
